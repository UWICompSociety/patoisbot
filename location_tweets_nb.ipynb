```python
class CustomStreamListener(tweepy.StreamListener):
    def on_status(self, status):
        #if 'manchester united' in status.text.lower():
        print (status.text)
        file = open("jamtweets.txt", "a+")
        file.write(status.text + "\n")
        file.close()

    def on_error(self, status_code):
        print('Encountered error with status code:', status_code)
        return True # Don't kill the stream

    def on_timeout(self):
        print('Timeout...')
        return True # Don't kill the stream
```

```python
sapi = tweepy.streaming.Stream(auth, CustomStreamListener())    
sapi.filter(locations=[-78.3377192858, 17.7011162379, -76.1996585761, 18.5242184514])
```

```python
new_lines = []
for line in lines:
    l = line.lower().strip()
    #print(line)
    sentence = re.sub(r"([?.!,])", r" \1 ", l)
    sentence = re.sub(r'[" "]+', " ", sentence)
    # replacing everything with space except (a-z, A-Z, ".", "?", "!", ",")
    sentence = re.sub(r"[^a-zA-Z?.!,]+", " ", sentence)
    sentence = re.sub(r'^https?:\/\/.*[\r\n]*', '', sentence, flags=re.MULTILINE)
    sentence = sentence.strip()
    new_lines.append(sentence)
```

```python
full_text = ''.join(new_lines)
```
